```python
# DeLTA outline overlays provided by Owen. To process these, I need to convert outlines to masks, which is fine because DeLTA always has at least 1px separation between masks.
# Note that the naming convention for the images published on the Cellpose website needed to be numeric, but thanks to the way I exported those images, they are read in in the correct order here. 
delta_dir = os.path.join(basedir,'DeLTA')
paths = sorted(Path(delta_dir).rglob("*.png"),key=getname)
dt_names = [str(p) for p in paths]

from scipy.ndimage.morphology import binary_fill_holes 
from scipy.ndimage import label
def process_dt(f):
    im = skimage.io.imread(f)/255
    outlines = np.logical_and(im[:,:,0]==1, im[:,:,1]!=1, im[:,:,2]==1)
    binary_masks = binary_fill_holes(outlines)
    masks,_ = label(binary_masks)
    return masks

masks_dt = [omnipose.utils.format_labels(omnipose.utils.clean_boundary(process_dt(f))) for f in dt_names]
```

```python
from scipy.optimize import linear_sum_assignment

per_cell = 1
J = len(IoU)
K = len(cats)
# y = [[]*K]*J #BADDDDDDDDDD! Need to check my other results to make sure it was done correctly, https://stackoverflow.com/questions/54673821/python-how-to-initialize-a-nested-list-with-empty-values-which-i-can-append-to 
y = [ [ [] for i in range(K) ] for i in range(J) ]
for j in range(J):
    for k in range(K):
        matched_iou = []
        mean_matched_iou = []
        for ind in indices[k]:
            iou = IoU[j][ind][1:,1:]
            th = 0
            n_min = min(iou.shape[0], iou.shape[1])
            costs = -(iou >= th).astype(float) - iou / (2*n_min)
            true_ind, pred_ind = linear_sum_assignment(costs)
            miou = iou[true_ind, pred_ind]
            matched_iou.append(miou)
            mean_matched_iou.append(np.sum(miou)/max(iou.shape[0], iou.shape[1]))
            
        y[j][k] = [m for sublist in matched_iou for m in sublist] if per_cell else mean_matched_iou
```
